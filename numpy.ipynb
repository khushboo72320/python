```python
#reshaping or flatterning
import numpy as np
flat_arr=np.arange(6)
print("original 1d array:",flat_arr)
matrix_2d=flat_arr.reshape(2,3)
print(matrix_2d)
flattend = matrix_2d.flatten()
print("\nflattened array back to 1d:",flattend)
```
