```python
#for showing x label or y label plt.figure giving figure size linewidth width of line label of line (giving name of line) for title we write plt.title
days=[1,2,3,4,5]
speed=[10,20,30,49,86]
plt.figure(figsize=(7,4))
plt.plot(days,speed,color='yellow',marker='*',linestyle='--',linewidth=2,label='hero speed')
plt.plot(speed,days,color='red',marker='o',linestyle='--',linewidth=2,label='revenue')
plt.title("SSpeed over days",fontsize=14,color='darkred')
plt.xlabel("Days of training",fontsize=12)
plt.ylabel("Flight speed(mph)",fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#bar plot used for compare the categories 
treats=['chockletes','candies','cookies','mashmalloes']
counts=[3,6,4,8,]
plt.figure(figsize=(7,4))
plt.bar(treats,counts,color='skyblue',edgecolor='darkblue')
plt.title("traetseaten by friends this week",fontsize=14,color='teal')
plt.xlabel("types of treats",fontsize=12)
plt.ylabel("total eaten counts",fontsize=12)
plt.show()
```

```python
#used for  check relation and clusters
east_cords=[3,4,5,6,7,8.7,4.5]
north_cords=[2,1,4.9,9.8,8,7,8]
plt.figure(figsize=(7,4))
plt.scatter(east_cords,north_cords,color='gold',marker='*',s=100,edgecolor='darkgoldenrod')
plt.title("treasure chest ;location on the map")
plt.xlabel("east-west coordinate",fontsize=12)
plt.ylabel("north-south coordinate",fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
np.random.seed(42)
candies=np.random.normal(loc=20,scale=4,size=30).astype(int)
df_candy=pd.DataFrame({"candies":candies})
print(df_candy)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(7,4))
sns.histplot(data=df_candy,x="candies",kde="True",color="purple",bins=10)
plt.title("candies in the box")
plt.xlabel("candies",fontsize=12)
plt.ylabel("candies is eaten",fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
frog_jumps=[1,2,3,4,5,19]
df_frogs=pd.DataFrame({"JumpDistance":frog_jumps})
print(df_frogs)
plt.figure(figsize=(7,4))
sns.boxplot(data=df_frogs,x="JumpDistance",color="lightgreen")
plt.title("frogs")
plt.xlabel("froms jump distance &outlier detection",fontsize=12,color='darkgreen')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
games_df=pd.DataFrame({
'playtime_hours':[10,50,80,100],
'popularity_score':[83,95,77,89],
'game_difficulty':[8,6,7,9]})
corr_matrix=games_df.corr()
print(corr_matrix)
plt.figure(figsize=(6,4.5))
sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",vmin=-1,vmax=1,linewidth=1)
plt.title("video game",fontsize=14)
plt.show()
```
